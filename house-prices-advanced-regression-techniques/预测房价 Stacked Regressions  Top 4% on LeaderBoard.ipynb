```python
#import 相关的包
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import matplotlib.pyplot as plt  # Matlab-style plotting
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
import warnings
def ignore_warn(*args, **kwargs):
    pass
warnings.warn = ignore_warn #ignore annoying warning (from sklearn and seaborn)


from scipy import stats
from scipy.stats import norm, skew #for some statistics


pd.set_option('display.float_format', lambda x: '{:.3f}'.format(x)) #Limiting floats output to 3 decimal points


from subprocess import check_output
```

```python
#pandas 读取数据
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#展示前5行
train.head()
```

```python
#查看数据的维度（shape）
print('训练集的维度是： {}'.format(train.shape))
print("测试集的维度是： {}".format(test.shape))

#保存ID列
train_ID = train['Id']
test_ID = test['Id']

#将ID列从原数据集去除，因为这些对于预测没有帮助

train.drop("Id", axis = 1, inplace=True) # inplace = True, 表示直接在原数据集上更改
test.drop("Id",axis =1, inplace=True)

print('去除 ID 后训练集的维度是： {}'.format(train.shape))
print("去除 ID 测试集的维度是： {}".format(test.shape))
```

# 数据处理

## 异常值/离群值 （Outliers）

```python
fig = sns.scatterplot(x='GrLivArea', y='SalePrice', data=train)
fig.set(xlim=(0,6000),ylim=(0,800000))
```

可以看到右下角有两个异常值，很大的GrLivArea 却对应着很低的房价，我们可以安全的将其删除

```python
#删除异常值

train = train.drop(train[(train['GrLivArea']>4000) & (train['SalePrice']<300000)].index)

#再次检查

fig = sns.scatterplot(x='GrLivArea', y='SalePrice', data=train)
fig.set(xlim=(0,6000),ylim=(0,800000))
```

### 注意
删除异常值是一个危险的事情！我们这里删除异常值是因为这两个很明显

## 房价是我们要预测的变量，我们先来分析一下它

```python
sns.distplot(train['SalePrice'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

可以看到数据右偏了，然而线性拟合喜欢呈 正态分布 的数据，我们需要做一些操作让这个变量分布的更加正态化一些

## 对目标变量进行对数转化

```python
#用numpy中的log1p函数 log(1+x)
train['SalePrice'] = np.log1p(train['SalePrice'])

#看一看转变之后的变量

sns.distplot(train['SalePrice'], fit=norm)

(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

### 可以看到数据分布的更加正态了
